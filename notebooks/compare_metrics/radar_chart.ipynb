```python
import plotly.graph_objects as go
# import streamlit as st # Bỏ comment nếu dùng Streamlit

def plot_radar_chart_optimized():
    # 5 trục của biểu đồ
    categories = ['NMI', 'ARI', 'Purity', 'Modularity', '1 - Conductance']

    # Khởi tạo biểu đồ
    fig = go.Figure()

    # ==========================================
    # CẤU HÌNH MÀU SẮC & ĐƯỜNG NÉT (Publication Quality)
    # Sử dụng bảng màu Qualitative để tạo sự khác biệt rõ nhất
    # Alpha (độ trong suốt) của vùng Fill được giảm xuống cực thấp (0.1)
    # ==========================================
    
    config = {
        'Leiden': {
            'line_color': 'rgba(31, 119, 180, 1)',  # Xanh dương đậm (Solid)
            'fill_color': 'rgba(31, 119, 180, 0.1)', # Trong suốt 90%
            'dash': 'solid',                        # Nét liền
            'marker': 'circle'                       # Nút hình tròn
        },
        'Louvain': {
            # Dùng màu Cam/Nâu đất để tương phản mạnh với Xanh nhưng vẫn trang trọng
            'line_color': 'rgba(217, 95, 2, 1)',    # Cam đất (Contrasting)
            'fill_color': 'rgba(217, 95, 2, 0.1)',  # Trong suốt 90%
            'dash': 'dash',                         # Nét đứt (Quan trọng để phân biệt khi trùng)
            'marker': 'square'                      # Nút hình vuông
        },
        'Spectral': {
            # Dùng màu Xanh lá cây mờ hơn để phân biệt với 2 thuật toán top trên
            'line_color': 'rgba(27, 158, 119, 1)',  # Xanh lá (Distinct)
            'fill_color': 'rgba(27, 158, 119, 0.1)', # Trong suốt 90%
            'dash': 'dot',                          # Nét chấm chấm
            'marker': 'diamond'                     # Nút hình kim cương
        }
    }

    # --- Thêm dữ liệu cho Leiden ---
    fig.add_trace(go.Scatterpolar(
        r=[0.8822,0.8138,0.837,0.8561,0.0654],
        theta=categories,
        fill='toself',
        name='Leiden (Top)',
        fillcolor=config['Leiden']['fill_color'],
        line=dict(
            color=config['Leiden']['line_color'], 
            width=3, # Đường viền dày hơn
            dash=config['Leiden']['dash']
        ),
        marker=dict(
            symbol=config['Leiden']['marker'], 
            size=8, # Nút to hơn để dễ nhìn nút trùng
            opacity=1
        )
    ))

    # --- Thêm dữ liệu cho Louvain (Gần trùng với Leiden) ---
    fig.add_trace(go.Scatterpolar(
        r=[0.8236,0.6679,0.7183,0.8387,0.0587],
        theta=categories,
        fill='toself',
        name='Louvain (Close Second)',
        fillcolor=config['Louvain']['fill_color'],
        line=dict(
            color=config['Louvain']['line_color'], 
            width=3, 
            dash=config['Louvain']['dash'] # Nét đứt sẽ lộ ra dưới nét liền của Leiden
        ),
        marker=dict(
            symbol=config['Louvain']['marker'], 
            size=8,
            opacity=1
        )
    ))

    # --- Thêm dữ liệu cho Spectral Clustering ---
    fig.add_trace(go.Scatterpolar(
        r=[0.544,0.1371,0.3775,0.5991,0.0248],
        theta=categories,
        fill='toself',
        name='Spectral Clustering',
        fillcolor=config['Spectral']['fill_color'],
        line=dict(
            color=config['Spectral']['line_color'], 
            width=2.5, 
            dash=config['Spectral']['dash']
        ),
        marker=dict(
            symbol=config['Spectral']['marker'], 
            size=8,
            opacity=1
        )
    ))

    # ==========================================
    # CẤU HÌNH LAYOUT CHUẨN BÁO KHOA HỌC
    # ==========================================
    fig.update_layout(
        title={
            'text': "<b>Figure 3: Comparative Analysis of Community Detection Efficiency</b>",
            'y': 0.97,
            'x': 0.5,
            'xanchor': 'center',
            'yanchor': 'top',
            'font': dict(size=20, family="Times New Roman", color="black")
        },
        font=dict(family="Times New Roman", size=14, color="black"),
        template="plotly_white", # Nền trắng sạch
        showlegend=True,
        legend=dict(
            orientation="h",       # Legend nằm ngang
            yanchor="bottom",
            y=-0.18,               # Đẩy legend xuống thấp để không che trục Y
            xanchor="center",
            x=0.5,
            font=dict(size=12),
            bgcolor='rgba(255,255,255,0.5)' # Nền legend hơi trong suốt
        ),
        polar=dict(
            bgcolor='white',
            radialaxis=dict(
                visible=True,
                range=[0, 1.0],         # Thang điểm 0-1
                gridcolor='#e0e0e0',  # Màu lưới mờ mảnh
                gridwidth=0.5,
                tickfont=dict(size=11, color='#666666')
            ),
            angularaxis=dict(
                gridcolor='#e0e0e0',
                gridwidth=0.5,
                tickfont=dict(size=13, color='black', family="Times New Roman")
            )
        ),
        # Tăng margin để các Metric name không bị cắt
        margin=dict(l=100, r=100, t=80, b=100)
    )

    # Hiển thị
    # st.plotly_chart(fig, use_container_width=True) # Dùng cho Streamlit
    fig.show() # Dùng cho Python script thường

if __name__ == "__main__":
    plot_radar_chart_optimized()
```
