```python
import plotly.graph_objects as go
import streamlit as st

def plot_bar_chart():
    # 1. Định nghĩa dữ liệu
    metrics = ['NMI', 'ARI', 'Purity', 'Modularity', 'Conductance']
    
    leiden_scores = [0.8822,0.8138,0.837,0.8561,0.0654]
    louvain_scores = [0.8236,0.6679,0.7183,0.8387,0.0587]
    spectral_scores = [0.544,0.1371,0.3775,0.5991,0.0248]

    # 2. Bảng màu chuyên nghiệp (Academic Palette)
    # Sử dụng các tông màu xanh đậm, xanh nhạt và xám bạc để tạo sự tương phản tốt khi in trắng đen
    colors = {
        'Leiden': '#1f4e79',      # Xanh Navy đậm
        'Louvain': '#2e75b6',     # Xanh trung bình
        'Spectral': '#bdd7ee'     # Xanh nhạt/Xám xanh
    }

    fig = go.Figure()

    # 3. Thêm các cột với định dạng văn bản trên đầu cột (text)
    algorithms = [
        ('Leiden', leiden_scores, colors['Leiden']),
        ('Louvain', louvain_scores, colors['Louvain']),
        ('Spectral Clustering', spectral_scores, colors['Spectral'])
    ]

    for name, scores, color in algorithms:
        fig.add_trace(go.Bar(
            name=name,
            x=metrics,
            y=scores,
            marker_color=color,
            text=[f'{s:.2f}' for s in scores], # Hiển thị 2 chữ số thập phân
            textposition='outside',           # Đặt số bên ngoài cột
            textfont=dict(size=10),
            error_y=None # Có thể thêm thanh sai số nếu có dữ liệu
        ))

    # 4. Tối ưu hóa Layout theo chuẩn báo cáo khoa học
    fig.update_layout(
        title={
            'text': "<b>Figure 1: Comparative Evaluation of Community Detection Algorithms</b>",
            'y': 0.95,
            'x': 0.5,
            'xanchor': 'center',
            'yanchor': 'top',
            'font': dict(size=18, family="Times New Roman")
        },
        xaxis_title="Evaluation Metrics",
        yaxis_title="Score Value",
        legend_title="Algorithms",
        font=dict(family="Times New Roman", size=14, color="black"), # Font chữ chuẩn bài báo
        
        template="plotly_white", # Nền trắng sạch sẽ
        barmode='group',
        bargap=0.2,    # Khoảng cách giữa các cụm
        bargroupgap=0.1, # Khoảng cách giữa các cột trong cụm
        
        # Cấu hình lề
        margin=dict(l=60, r=30, t=80, b=60),
        
        # Tùy chỉnh trục tọa độ
        yaxis=dict(
            range=[0, 1.1], # Để khoảng trống cho label trên đầu cột
            gridcolor='lightgrey', # Đường kẻ mờ
            zerolinecolor='black'
        ),
        xaxis=dict(
            linecolor='black',
            showgrid=False
        )
    )

    # Hiển thị
    # st.plotly_chart(fig, use_container_width=True) 
    fig.show()

if __name__ == "__main__":
    plot_bar_chart()
```
